```python
df = pd.read_csv('2020playerstats.csv')
```

```python
#Cleaning the data 
mapping = {df.columns[0]:'Index'}
df = df.rename(columns=mapping)
```

```python
#Exploring The Data 
plt.figure(figsize=(12,5))
plt.xticks(rotation = 45)
sns.lineplot(x=df['FantPos'], y=df['FantPt'], color='r', )
#WR and RB make up about the same # of Pts per season on average while QB's make up the highest Pts per season for most any team
#Insight: Picking a QB may be less important in early rounds as the average performance is higher 
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation = 45)
sns.barplot(x=df['Tm'],y=df['Age'], color='r')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation = 45)
sns.scatterplot(x=df['Y/A'], y=df['TD.1'], color='r')
```
